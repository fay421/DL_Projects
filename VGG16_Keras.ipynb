## VGG16 by Keras

## Import Libraries

```
from keras.applications import VGG16
from keras.applications.vgg16 import preprocess_input, decode_predictions
from keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
```

```
pre_trained_model = VGG16(weights='imagenet')
```

# Show the Loaded Image

```
path_img = '/content/Camel.jpg'
img = load_img(path_img)
plt.imshow(img);
```

## Preprocessing Phases:

#1- Resize the Image (224,224,3)

```
img = load_img(path_img,target_size=(224,224,3))
plt.imshow(img);
```

#2- Convert to array

```
arr_img = img_to_array(img)
print('array_image shape: ', arr_img.shape)
```

#3- Add the dummy dimension

```
arr_img =np.expand_dims(arr_img,axis=0)
print('array_image shape after add first dim: ', arr_img.shape)
```

## Predict the Image

```
vgg_img =preprocess_input(arr_img)
```

```
y_pred=pre_trained_model.predict(vgg_img)
```

```
print(y_pred)  # 1000 classes
```

## Decode the Classes

```
prediction = decode_predictions(y_pred,top=5)
```

```
print(prediction)
```

## More Examples

```
path_img = ['/content/Frog.jpeg', '/content/Owl.webp']
# Load and display the image
for i in range(len(path_img)):
  # Process the image for prediction
  img = load_img(path_img[i], target_size=(224, 224))
  arr_img = img_to_array(img)
  arr_img = np.expand_dims(arr_img, axis=0)
  vgg_img = preprocess_input(arr_img)
  y_pred = pre_trained_model.predict(vgg_img)
  prediction = decode_predictions(y_pred, top=1)
  print(prediction)
  plt.imshow(img)
  plt.show()
```
